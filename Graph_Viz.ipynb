```python
import networkx as NX
import gzip, os, sys
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = NX.Graph()

nsteps = 10000
for fn in os.listdir('.'):
    if fn[:6] == 'time_0':
        with open(fn, 'r') as f:
            for i, l in enumerate(f):
                if i == nsteps:
                    break
                v = l.strip().split()
                G.add_edge(v[0], v[1])

G_symm = NX.Graph()

for fn in os.listdir('.'):
    if fn[:9] == 'time_symm':
        with gzip.open(fn, 'rb') as f:
            for i, l in enumerate(f):
                if i == nsteps:
                    break
                v = l.strip().split()
                G_symm.add_edge(v[0], v[1])
                
```

```python
pos = NX.spring_layout(G)
pos_symm = NX.spring_layout(G_symm)
```

```python
f=plt.figure()

ax=f.add_subplot(1,2,1)
NX.draw(G, pos=pos, hold=True, ax=ax)

ax_s=f.add_subplot(1,2,2)
NX.draw(G_symm, pos=pos_symm, hold=True, ax=ax_s)

f.set_size_inches(40,15)
```

```python
Ass = NX.assortativity.average_neighbor_degree(G)
Ass = [Ass[k] for k in sorted(G.nodes())]
Deg = [G.degree(k) for k in sorted(G.nodes())]

Ass_S = NX.assortativity.average_neighbor_degree(G_symm)
Ass_S = [Ass_S[k] for k in sorted(G_symm.nodes())]
Deg_S = [G_symm.degree(k) for k in sorted(G_symm.nodes())]
```

```python
plt.plot(Deg, Ass, '.b')
plt.plot(Deg_S, Ass_S, '*r')
```

```python
print NX.cluster.average_clustering(G)

print NX.cluster.average_clustering(G_symm)
```
